```python
import numpy as np
import pandas as pd
import pdb

extracted_segments = pd.read_pickle('/Users/matthewashman/github/MasterProject2018/EPDataAnalysis/Final Report/extracted_segments.pkl')
with open('/Users/matthewashman/github/MasterProject2018/EPDataAnalysis/Final Report/final_segment_labels.csv', 'r') as file_in:
    labels = pd.read_csv(file_in)
```

```python
labels = labels.dropna(how='all')
extracted_segments['Label'] = np.nan
for idx, row in labels.iterrows():
    coupling_interval = str(int(row['CI']))
    patient_num = str(row['Patient'])
    patient_type = row['Type']
    patient_type = patient_type.lower()
    # Add labels to dataframe
    extracted_segments.loc[(extracted_segments['Type']==patient_type) & (extracted_segments['Patient']==patient_num) & (extracted_segments['Channel']=='CS1-2') & (extracted_segments['Coupling Interval']==coupling_interval) & (extracted_segments['S1/S2']=='S2'), 'Label'] = str(row['CS1-2'])
    extracted_segments.loc[(extracted_segments['Type']==patient_type) & (extracted_segments['Patient']==patient_num) & (extracted_segments['Channel']=='CS3-4') & (extracted_segments['Coupling Interval']==coupling_interval) & (extracted_segments['S1/S2']=='S2'), 'Label'] = str(row['CS3-4'])
    extracted_segments.loc[(extracted_segments['Type']==patient_type) & (extracted_segments['Patient']==patient_num) & (extracted_segments['Channel']=='CS5-6') & (extracted_segments['Coupling Interval']==coupling_interval) & (extracted_segments['S1/S2']=='S2'), 'Label'] = str(row['CS5-6'])
```
