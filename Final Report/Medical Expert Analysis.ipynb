```python
import sys
sys.path.insert(0, '/Users/matthewashman/github/MasterProject2018')

# Import necessary modules. Set settings. Import data.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import HTML

# Miscelaneous
from IPython.display import display, clear_output
import pdb
import re
import warnings
warnings.filterwarnings('ignore')

plt.style.use('default')

with open('/Users/matthewashman/github/MasterProject2018/EPDataAnalysis/Final Report/professional_analysis_results.csv', 'r') as file_in:
    expert_analysis_file = pd.read_csv(file_in)
```

```python
expert_analysis_file.head()
```

```python
expert_analysis_list = []
for i, row in expert_analysis_file.iterrows():
    patient_dict = {}
    patient_type = " ".join(re.findall("[a-zA-Z]+", row['Patient']))
    patient_type = patient_type.lower()
    patient = " ".join(re.findall("[0-9]+", row['Patient']))
    
    patient_dict['Type'] = patient_type
    patient_dict['Patient'] = patient
    patient_dict['Conduction Delay'] = [row['Conduction Delay ' + str(i)] for i in range(1,4)]
    patient_dict['Fractionation'] = [row['Fractionation ' + str(i)] for i in range(1,4)]
    patient_dict['Stop Early'] = [(row['Stop Early ' + str(i)].upper() == 'Y') for i in range(1,4)]
    patient_dict['S1/S2 Interval'] = [row['Interval ' + str(i)] for i in range(1,4)]
    
    expert_analysis_list.append(patient_dict)
        
```

```python
expert_analysis_pd = pd.DataFrame(expert_analysis_list)
af_expert_analysis = expert_analysis_pd[expert_analysis_pd['Type']=='af']
non_af_expert_analysis = expert_analysis_pd[expert_analysis_pd['Type']!='af']
```

```python
af_fractionation = np.array([np.array(xi) for xi in af_expert_analysis['Fractionation'].values])
non_af_fractionation = np.array([np.array(xi) for xi in non_af_expert_analysis['Fractionation'].values])

% matplotlib qt

fig = plt.figure(figsize=(10,3), dpi=80)
for x in af_fractionation:
    x = x + np.random.normal(loc=0, scale=0.05, size=len(x))
    y = np.linspace(0.75, 1.25, len(x))
    plt.plot(x, y, 'ro')
    
for x in non_af_fractionation:
    y = np.linspace(0, 0.5, len(x))
    plt.plot(x, y, 'go')
    
plt.show()
```
